```python
#Удаление стоп-слов
#Импорт стоп-слов из бибилиотеки nltk
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('russian'))

# функция, удаляющая стопслова из текстов
def stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
# Удаление эмодзи
def emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
#Удаление URL
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```

```python
#Удаление HTML-тегов
from bs4 import BeautifulSoup

def html(text):
    return BeautifulSoup(text, "lxml").text
```

```python
#токенезация и лемматизация при помощи NLTK
from pymorphy3 import MorphAnalyzer
from nltk import sent_tokenize, word_tokenize, regexp_tokenize

def tokenize_lemmas(sent, pat=r"(?u)\b\w\w+\b", morph=MorphAnalyzer()):
    return [morph.parse(tok)[0].normal_form 
            for tok in regexp_tokenize(sent, pat)]
```

```python
#удаление часто встречающиеся слов в корпусе при помощи tf-idf
from collections import Counter
cnt = Counter()
freq = set([w for (w, wc) in cnt.most_common(10)])
# функция удаления слов
def freqwords(text):
    return " ".join([word for word in str(text).split() if word not 
in freq])
```

```python
#в новой колонке text_ready будут обработанные отзывы

df['text_ready']  = df['review'].str.lower() #убирает заглавные буквы
df['text_ready'] = df['text_ready'].str.replace(r'\d+', '', regex=True) #удаление цифр 
df['text_ready'] = df['text_ready'].str.replace('[^\\w\\s]','') #пунктуация 
df["text_ready"] = df["text_ready"].apply(stopwords) #удаляет стоп-слова (ф-я вышк)
df['text_ready'] = df['text_ready'].apply(emoji) 
df['text_ready'] = df['text_ready'].apply(remove_urls)
df['text_ready'] = df['text_ready'].apply(html)
df["text_ready"] = df["text_ready"].map(lambda x: " ".join(tokenize_lemmas(x))) #токенезация и лемматизация
df['text_ready'] = df['text_ready'].apply(freqwords)
df['text_ready']
```

```python
#Удаление низкочастотных слов
freq = pd.Series(' '.join(df['text_ready']).split()).value_counts()[-10:] # 10 rare words
freq = list(freq.index)
df['text_ready'] = df['text_ready'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df['text_ready']
```
